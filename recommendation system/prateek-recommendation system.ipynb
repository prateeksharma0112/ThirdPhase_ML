## Data Preprocessing

```python
### Important Parameters which required movie_id, title, overview, genres, keywords, cast, crew #######

movies = movies[["movie_id", "title", "overview", "genres", "keywords", "cast", "crew"]]
```

```python
# Now convert Structures of Data in the Dataframe for genere, keywords, cast and crew. Put it into List.
```

```python
import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    
    return L
```

```python
# For Genres 
movies['genres'] = movies['genres'].apply(convert)
```

```python
# For Keywords
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
# For Cast
movies['cast']
```

```python
def convert3(text):
    L = []
    c = 0
    for i in ast.literal_eval(text):
        if c<3:
            L.append(i['name'])
            c +=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
# For Crew
movies['crew']
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
         
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
# Some correctness in our dataframe
def removespaces(text):
    L1 = []
    for i in text:
        L1.append(i.replace(" ", ""))
    return L1    
```

```python
movies['genres']= movies['genres'].apply(removespaces)
```

```python
movies['keywords'] = movies['keywords'].apply(removespaces)
```

```python
movies['cast'] = movies['cast'].apply(removespaces)
```

```python
movies['crew'] = movies['crew'].apply(removespaces)
```

```python
#### make overview, from string into list
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
################## Concatenation of genres, keywords, cast, crew into tags ######################
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(1)
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_df
```

```python
## convert list into string in tags

new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

```python
## convert it in to lowercase

new_df['tags'] = new_df['tags'].apply(lambda x: x.lower())
```

```python
new_df['tags'][0]
```

```python
## Apply Stemming

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    S = []
    for i in text.split():
        S.append(ps.stem(i))
    S = " ".join(S)
    return S    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

## Convert Text into Vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names()
```

```python
# Cosine Similarity [between 0 and 1 ]

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Har Movie ka 4806 movies ke saath distance
similarity = cosine_similarity(vector)
```

```python
# function which return 5 Similar Movie

def movie_recommend(movie):
    movie_idx = new_df[new_df['title'] == movie].index[0]
#     distances = similarity(movie_idx)
    movies_lst = sorted(list(enumerate(similarity[movie_idx])), reverse=True, key=lambda x:x[1])[1:11]
    
    for i in movies_lst:
#       print(i)
#       print(i[0])
        print(new_df['title'][i[0]])
```
